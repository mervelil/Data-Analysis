```python
import numpy as np
bins=np.linspace(df2['tip'].min(), df2['tip'].max(),4)
labels=['small','medium','large']
```

```python
df2['tip']=pd.cut(df2['tip'], bins=bins, labels=labels)
```

```python
df2.tip.unique()
```

```python
# Using KBinsDiscretizer
scaler=skp.KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='kmeans')
digital_labels=scaler.fit_transform(df2['tip'].values.reshape(-1,1))
digital_labels=digital_labels.astype(str)
digital_labels[digital_labels=='0.0']='small'
digital_labels[digital_labels=='1.0']='medium'
digital_labels[digital_labels=='2.0']='high'
```

```python
df2['tip_as_cat']=digital_labels
```
