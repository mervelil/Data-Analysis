Using the data frame countries:

- Drop columns with missing values 
- Limit to countries with populations over 20 million
- Reset index to start from zero and count over remaining countries, and remove additional columns.
- Add a boolean valued column that is true for countries with coastlines.
- Calculate mean populations
- Calculate the sum of all populations in Asia
- Find the number of land-locked countries.
- Find the mean of ('population', 'area', 'coastline') of different 'continents'.
- Find the minimum and maximum of the 'life expectancy' for costal and landlocked countries.
- How many coastal and landlocked countries are in each continent? Determine the mean deathrate for coastal and landlocked countries are in each continent.
- Create a new column measuring population density.  Determine the mean population density of all the countries.  Create a new column 'dense' that is True for countries with higher than average density and False for lower than average density.
- For each category of 'dense' compute the mean birthrate and the difference between the maximal birthrate and the minimal birthrate.

```python
import pandas as pd
df=pd.read_csv('countries.csv', sep=';')
```

```python
#Display top ten rows of the data
df.head(10)
```

```python
#Choose the following columns from the data: 'Country','Continent','Population','Area','Coastline','Currency','Birthrate','Deathrate','Life expectancy'.
df=df[['Country','Continent','Population','Area','Coastline','Currency','Birthrate','Deathrate','Life expectancy']]
df.head()
```

```python
#Change the column names to lowercase
df.columns=df.columns.str.lower()
df.head()
```

```python
#Find number of missing values
df.isna().sum()
#Drop columns with missing values
df=df.dropna()
```

```python
#Limit to countries with populations over 20 million
df=df.loc[df.population>=20e6,:]
```

```python
#Reset index to start from zero and count over remaining countries, and remove additional columns.
df=df.reset_index().drop('index',axis=1)
df.head()
```

```python
#Add a boolean valued column that is true for countries with coastlines.
df['coastal']=df['coastline']>0
df.head()
```

```python
#Calculate the sum of all populations in Asia
df.loc[df.continent=='Asia','population'].sum()

#Find the number of land-locked countries.
len(df.loc[df.coastal==False,:])
```

```python
#Find mean of ('population', 'area', 'coastline') of different 'continents'.
df.groupby('continent').mean()[['population', 'area', 'coastline']]
```

```python
#Find the minimum and maximum of the 'life expectancy' for costal and landlocked countries.
df.groupby('coastal').agg(['min','max'])['life expectancy']
```

```python
#Calc mean populations in each continent
df.groupby('continent').mean()[['population']]
```

```python
#Create a new column measuring population density.  Determine the mean population density of all the countries.  
```

```python
df['density']=df['population']/df['area']
df.head()
```

```python
# Create a new column 'dense' that is True for countries with higher than average density and False for lower than average density.
df['dense']=df['density']>df['density'].mean()
```

```python
df.head(10)
```

```python
def diff_max_min(col):
    return col.max() - col.min()
```

```python
#For each category of 'dense' compute the mean birthrate and the difference between the maximal birthrate and the minimal birthrate.
df.groupby('dense').agg(['mean', diff_max_min])['birthrate']
```
